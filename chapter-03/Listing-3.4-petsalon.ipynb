```python
"""PyCharm (No LangChain) â€” OpenRouter chat (replace Azure completions, no main())"""

# ============================================================================
# IMPORTS
# ============================================================================
import os
from dotenv import load_dotenv
from openai import OpenAI

# ============================================================================
# ENVIRONMENT SETUP (fail-fast)
# ============================================================================
load_dotenv()
OPENROUTER_API_KEY  = os.getenv("OPENROUTER_API_KEY")
OPENROUTER_BASE_URL = os.getenv("OPENROUTER_BASE_URL")

if not OPENROUTER_API_KEY:
    raise RuntimeError("Missing OPENROUTER_API_KEY in .env")
if not OPENROUTER_BASE_URL:
    raise RuntimeError("Missing OPENROUTER_BASE_URL in .env")

# ============================================================================
# CLIENT INIT
# ============================================================================
client = OpenAI(
    api_key=OPENROUTER_API_KEY,
    base_url=OPENROUTER_BASE_URL,
)

# ============================================================================
# PROMPT
# ============================================================================
prompt = (
    "Suggest three names for a new pet salon business. "
    "The generated name ideas should evoke positive emotions and these key features: "
    "Professional, Friendly, Personalized Service. Respond as 3 bullet points."
)

# ============================================================================
# CHAT COMPLETION
# ============================================================================
resp = client.chat.completions.create(
    model="openai/gpt-4o",
    temperature=0.7,
    max_tokens=100,
    messages=[
        {"role": "system", "content": "You generate short, catchy brand names."},
        {"role": "user",    "content": prompt},
    ],
)

# ============================================================================
# OUTPUT
# ============================================================================
print("Prompt:", prompt)
print("Response:\n", resp.choices[0].message.content.strip())
```
